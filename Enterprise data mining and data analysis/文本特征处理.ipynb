**尝试对经营范围进行处理**

采用TF-IDF方法进行向量分析，完成关键词提取

之后采用LDA主题模型完成主题分析

生成可视化图像

```python
import pandas as pd
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from collections import Counter
import matplotlib.pyplot as plt

# 配置matplotlib以支持中文字符
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为中文宋体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
csvFile='./data'+os.sep+'data01.csv'
data = pd.read_csv(csvFile,encoding='utf-8')
# print(data) #全部输出
print("\n输出数据前5条")
print(data.head()) #只输出前五条
```

```python
# 提取“公司经营范围”列
business_scope_column = data['BS']

# 初始化TF-IDF向量化器
tfidf_vectorizer = TfidfVectorizer()

# 将文本转换为TF-IDF特征矩阵
tfidf_matrix = tfidf_vectorizer.fit_transform(business_scope_column)

# 获取特征词
feature_names = tfidf_vectorizer.get_feature_names_out()

# 计算每个词语的平均TF-IDF权重
average_tfidf_weights = tfidf_matrix.mean(axis=0).A1

# 创建词语与对应平均TF-IDF权重的字典
word_tfidf_dict = dict(zip(feature_names, average_tfidf_weights))

# 按照平均TF-IDF权重对字典进行排序
sorted_word_tfidf = sorted(word_tfidf_dict.items(), key=lambda x: x[1], reverse=True)

# 提取排序后的词语和对应的权重
words = [word[0] for word in sorted_word_tfidf]
weights = [word[1] for word in sorted_word_tfidf]

# 绘制条形图
plt.figure(figsize=(10, 6))
plt.barh(words[:20], weights[:20], color='skyblue')
plt.xlabel('Average TF-IDF Weight')
plt.ylabel('Word')
plt.title('Top 20 Words by Average TF-IDF Weight')
plt.gca().invert_yaxis()  # 反转y轴，使得权重最高的词在顶部显示

# 保存关键词提取结果
keywords_df = pd.DataFrame({'Word': words, 'Average TF-IDF Weight': weights})
keywords_df.to_csv('out/keywords_tfidf.csv', encoding='utf_8_sig', index = False)

# 保存图像
plt.savefig('image/tfidf_keywords.png')

plt.show()
```

经过分析文本，并提取关键词，发现企业经营范围主要以技术服务、技术咨询、技术转让为主，“技术”，“法”等在其中占了很大的分量。

```python
# 读取关键词提取结果文件
keywords_df = pd.read_csv('out/keywords_tfidf.csv')

# 假设关键词存储在名为'Word'的列中
corpus = keywords_df['Word'].values.tolist()

# 文本向量化
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

# 拟合LDA主题模型
lda_model = LatentDirichletAllocation(n_components=5, random_state=42)  # 假设有5个主题
lda_model.fit(X)

# 输出每个主题的关键词
feature_names = vectorizer.get_feature_names_out()
topics_keywords = []
for idx, topic in enumerate(lda_model.components_):
    keywords = [feature_names[i] for i in topic.argsort()[-10:]]  # 输出每个主题的前10个关键词
    topics_keywords.append(keywords)

# 提取文档的主题分布
doc_topic_distribution = lda_model.transform(X)

# 输出每个主题的关键词
feature_names = vectorizer.get_feature_names_out()
for idx, topic in enumerate(lda_model.components_):
    print(f"Topic {idx+1}:")
    print([feature_names[i] for i in topic.argsort()[-10:]])  # 输出每个主题的前10个关键词
    print()

# 保存原始输出到文件
with open('out/original_lda_output.txt', 'w') as f:
    print("Original LDA Output:", file=f)
    for idx, topic in enumerate(lda_model.components_):
        print(f"Topic {idx+1}:", file=f)
        print([feature_names[i] for i in topic.argsort()[-10:]], file=f)
        print(file=f)
    print(doc_topic_distribution, file=f)

# 保存主题和关键词到CSV文件
topics_df = pd.DataFrame(topics_keywords, columns=[f"Keyword {i+1}" for i in range(10)])
topics_df.to_csv('out/lda_topics_keywords.csv', index=False, encoding='utf-8-sig')

# 保存文档的主题分布到CSV文件
doc_topic_df = pd.DataFrame(doc_topic_distribution, columns=[f"Topic {i+1}" for i in range(5)])
doc_topic_df.to_csv('out/doc_topic_distribution.csv', index=False, encoding='utf-8-sig')

print("存储完成！")
```

Topic 1: 这个主题似乎涉及到科技领域，包括涂料研发、智能光电系统、大数据科技等方面。可能是涉及到新材料、智能技术和数据技术等领域的公司经营范围。

Topic 2: 这个主题涉及到规划和技术服务，如城乡规划、办公设备服务、水资源管理等。可能是提供规划咨询、技术服务和城市基础设施建设方面的公司。

Topic 3: 这个主题涉及到工程和技术服务，如农林工程、水电安装维护、生物科技研发等。可能是提供工程设计、技术研发和设备安装维护等方面的公司。

Topic 4: 这个主题涉及到通信、网络、品牌推广等领域，如通信网络设计、品牌推广、房地产咨询等。可能是通信、网络、品牌营销等行业的公司。

Topic 5: 这个主题涉及到环保和健康领域，如卫生监测、环保许可证、固体废物处理等。可能是从事环保、健康检测等方面的公司。
